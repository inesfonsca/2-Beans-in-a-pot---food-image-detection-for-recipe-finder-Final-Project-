```python
import pandas as pd
import openpyxl
from IPython.display import display, HTML
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
def recipe_Table():
# Read all sheets from the Excel file
    excel_file = pd.ExcelFile(r"C:\Users\ines_\Documents\Final Project\24Kitchen RECEITAS -  LINKS SITE_2.xlsx")

# Get a list of sheet names
    sheet_names = excel_file.sheet_names
    sheets = {"Name": [], "DataFrame": []}


# Read each sheet into a separate DataFrame
    dataframes = {}
    for sheet_name in sheet_names:
        df = excel_file.parse(sheet_name)
        dataframes[sheet_name] = df
        sheets["Name"].append(sheet_name)
        sheets["DataFrame"].append(df)

# Set column headers for each DataFrame
    for i, df in enumerate(sheets["DataFrame"]):
        sheets["DataFrame"][i].columns = df.iloc[0]
        sheets["DataFrame"][i] = df.iloc[1:].reset_index(drop=True)

    return sheets
```

```python
# Call the recipe_Table() function
recipe_data = recipe_Table()

# Select specific columns and create the DataFrame
ALABARRIOSS1 = pd.DataFrame(recipe_data["DataFrame"][0], columns=["RECEITA", "URL SITE", "YOUTUBE"])
ComTradição_1 = pd.DataFrame(recipe_data["DataFrame"][1], columns=["RECEITA", "URL SITE", "YOUTUBE"])
ComTradição_2= pd.DataFrame(recipe_data["DataFrame"][2], columns=["RECEITA", "URL SITE", "YOUTUBE"])
ComTradição_3= pd.DataFrame(recipe_data["DataFrame"][3], columns=["RECEITA", "URL SITE", "YOUTUBE"])
ComTradição_6= pd.DataFrame(recipe_data["DataFrame"][4], columns=["RECEITA", "URL SITE", "YOUTUBE"])
OdaJoana_1= pd.DataFrame(recipe_data["DataFrame"][5], columns=["RECEITA", "URL SITE", "YOUTUBE"])
OdaJoana_2= pd.DataFrame(recipe_data["DataFrame"][6], columns=["RECEITA", "URL SITE", "YOUTUBE"])
TiaCatia_8= pd.DataFrame(recipe_data["DataFrame"][7], columns=["RECEITA", "URL SITE", "YOUTUBE"])
TiaCatia_9= pd.DataFrame(recipe_data["DataFrame"][8], columns=["RECEITA", "URL SITE", "YOUTUBE"])
TiaCatia_12= pd.DataFrame(recipe_data["DataFrame"][9], columns=["RECEITA", "URL SITE", "YOUTUBE"])
ReceitasNatalTiaCatia= pd.DataFrame(recipe_data["DataFrame"][10], columns=["RECEITA", "URL SITE", "YOUTUBE"])
CadernosdaFilipa_1= pd.DataFrame(recipe_data["DataFrame"][11], columns=["RECEITA", "URL SITE", "YOUTUBE"])
CozinhaComTwist_1= pd.DataFrame(recipe_data["DataFrame"][12], columns=["RECEITA", "URL SITE", "YOUTUBE"])
DocesDoOfício_1= pd.DataFrame(recipe_data["DataFrame"][13], columns=["RECEITA", "URL SITE", "YOUTUBE"])
ReceitasNatal= pd.DataFrame(recipe_data["DataFrame"][14], columns=["RECEITA", "URL SITE", "YOUTUBE"])
```

```python
# Concatenate the DataFrames vertically
concatenated_df = pd.concat([ALABARRIOSS1, ComTradição_1, ComTradição_2, ComTradição_3, ComTradição_6,
                            OdaJoana_1, OdaJoana_2, TiaCatia_8, TiaCatia_9, TiaCatia_12,
                            ReceitasNatalTiaCatia, CadernosdaFilipa_1, CozinhaComTwist_1,
                            DocesDoOfício_1, ReceitasNatal], ignore_index=True)

```

```python
pd.set_option('display.max_rows', None)
# Display the concatenated DataFrame
concatenated_df.dropna(subset=['RECEITA'], inplace=True)
concatenated_df.reset_index(drop=True)
```

```python
final_df = pd.merge(Recipes, concatenated_df, on='RECEITA', how='inner')
```
