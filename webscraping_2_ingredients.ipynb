```python
#Importing the necessary libraries

import pandas as pd
import openpyxl
```

```python
#Extracting from an excel file the names of the recipes and their links
def recipe_Table():
# Read all sheets from the Excel file
    excel_file = pd.ExcelFile(r"C:\Users\ines_\Documents\Final Project\24Kitchen RECEITAS -  LINKS SITE_1.xlsx")

# Get a list of sheet names
    sheet_names = excel_file.sheet_names
    sheets = {"Name": [], "DataFrame": []}


# Read each sheet into a separate DataFrame
    dataframes = {}
    for sheet_name in sheet_names:
        df = excel_file.parse(sheet_name)
        dataframes[sheet_name] = df
        sheets["Name"].append(sheet_name)
        sheets["DataFrame"].append(df)

# Set column headers for each DataFrame
    for i, df in enumerate(sheets["DataFrame"]):
        sheets["DataFrame"][i].columns = df.iloc[0]
        sheets["DataFrame"][i] = df.iloc[1:].reset_index(drop=True)

    return sheets
```

```python
#Extracting the hyperlinks and links from the excel file

import openpyxl

def hyperlinks4():
    # Load the Excel file
    workbook = openpyxl.load_workbook(r"C:\Users\ines_\Documents\Final Project\24Kitchen RECEITAS -  LINKS SITE.xlsx")

    # Initialize a dictionary to store the links for each sheet
    sheet_links = {}

    # Iterate over all sheets
    for sheet_name in workbook.sheetnames:
        sheet = workbook[sheet_name]

        # Skip hidden sheets
        if sheet.sheet_state == "hidden":
            continue

        # Define the range of cells to check for hyperlinks
        start_row = 1
        end_row = sheet.max_row
        start_column = 1
        end_column = sheet.max_column

        # Initialize a list to store the links for the current sheet
        hyperlinks = []

        # Iterate over the cells within the specified range
        for row in sheet.iter_rows(min_row=start_row, max_row=end_row, min_col=start_column, max_col=end_column):
            for cell in row:
                if cell.hyperlink is not None and "24kitchen" in cell.hyperlink.target.lower():
                    hyperlink = cell.hyperlink.target
                    hyperlinks.append(hyperlink)

        # Store the links for the current sheet in the dictionary
        sheet_links[sheet_name] = hyperlinks

    # Print the links for each sheet
    for sheet_name, links in sheet_links.items():
        print("Sheet:", sheet_name)
        for link in links:
            print(link)
        print()
```

```python
#Extracting the ingredients for each link

import openpyxl
import requests
from bs4 import BeautifulSoup
import pandas as pd

def hyperlinks():
    # Load the Excel file
    workbook = openpyxl.load_workbook(r"C:\Users\ines_\Documents\Final Project\24Kitchen RECEITAS -  LINKS SITE.xlsx")

    # Initialize a dictionary to store the links for each sheet
    sheet_links = {}

    # Iterate over all sheets
    for sheet_name in workbook.sheetnames:
        sheet = workbook[sheet_name]

        # Skip hidden sheets
        if sheet.sheet_state == "hidden":
            continue

        # Define the range of cells to check for hyperlinks
        start_row = 1
        end_row = sheet.max_row
        start_column = 1
        end_column = sheet.max_column

        # Initialize a list to store the links for the current sheet
        hyperlinks = []

        # Iterate over the cells within the specified range
        for row in sheet.iter_rows(min_row=start_row, max_row=end_row, min_col=start_column, max_col=end_column):
            for cell in row:
                if cell.hyperlink is not None and "24kitchen" in cell.hyperlink.target.lower():
                    hyperlink = cell.hyperlink.target
                    hyperlinks.append(hyperlink)

        # Store the links for the current sheet in the dictionary
        sheet_links[sheet_name] = hyperlinks

    # Fetch ingredients from hyperlinked web pages and store in a DataFrame
    data = []
    for sheet_name, hyperlinks in sheet_links.items():
        for hyperlink in hyperlinks:
            response = requests.get(hyperlink)
            soup = BeautifulSoup(response.content, 'html.parser')

            ingredients_list = soup.find('ul', class_='recipe-ingredients__list')
            if ingredients_list:
                ingredients = [li.text.strip() for li in ingredients_list.find_all('li')]
                ingredients = ', '.join(ingredients)

                data.append({'Sheet': sheet_name, 'Ingredients': ingredients})

    # Create a pandas DataFrame from the collected data
    df = pd.DataFrame(data)
    
    return df
```
